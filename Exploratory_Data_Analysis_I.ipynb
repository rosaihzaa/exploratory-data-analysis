## Analysis Univariate

```python
# outliers check
plt.figure(figsize=(6, 3))

for column in range(0,len(numeric)):
  sns.boxplot(y=lung[numeric[column]], palette='flare')
  plt.title("Boxplot of Age")
  plt.tight_layout()
```

There are a few outliers in the AGE column boxplot, indicating that there are ages that have an extreme distance from other ages.

```python
# Displays distribution and histogram
plt.figure(figsize=(6, 3))

for column in range(0, len(numeric)):
  sns.histplot(x=lung[numeric[column]], palette='flare', kde=True)
  plt.xlabel(numeric[column])
  plt.title('Distributiion of Age')
  plt.tight_layout()
```

The data distribution is close to normal with the peak data around the age of 60 years indicating that the majority of individuals in the dataset are in that range.

## Deep Dive EDA

```python
lung.head()
```

Gender manakah yang paling banyak menderita lung cancer dan perokok?

```python
lung.columns
```

```python
gender_smoke = (
    lung.groupby('GENDER')
    .agg(lung_cancer = ('LUNG_CANCER', 'count'),
         smoker = ('SMOKING', 'count')
         )
    .reset_index()
    .sort_values('lung_cancer', ascending=False)
)
gender_smoke
```

Individuals in this data are men, the highest group suffering from lung cancer, and most of them are smokers.
